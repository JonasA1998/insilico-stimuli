# Gabors

```python
gabor_set = GaborSet(
    canvas_size=[64, 36],
    center_range=[30, 31, 15, 16], # All Gabors centered at (30, 15)
    sizes=[15, 30],
    spatial_frequencies=[0.5,2],
    contrasts=[0.5,1],
    orientations=2, # Two linearly spaced orientations; [2] would be orientation of 2 radians
    phases=2, # Same as orinetations
    grey_level=-0.1,
    #pixel_boundaries=[-1,1],
    #eccentricities=[0],
    #locations=[[20,20],[40,40]],
    #relative_sf=True
)
```

```python
gabor_set.params_dict_from_idx(1)
```

```python
plt.figure(figsize=(20, 5))

for i, img in enumerate(gabor_set.images()):
    plt.subplot(8, 4, i + 1)
    plt.imshow(img, cmap='gray', vmin=-1, vmax=1)
    plt.axis('off')
```
