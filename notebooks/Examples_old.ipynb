```python
%load_ext autoreload
%autoreload 2

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib.pyplot as plt

from stimuli import GaborSet, DiffOfGaussians, CenterSurround, PlaidsSet
```

# Gabors

```python
gabor_set = GaborSet(
    canvas_size=[64, 36],
    center_range=[30, 31, 15, 16], # All Gabors centered at (30, 15)
    sizes=[10, 30],
    spatial_frequencies=[0.1],
    contrasts=[2],
    orientations=2, # Two linearly spaced orientations; [2] would be orientation of 2 radians
    phases=2 # Same as orinetations
)
```

```python
gabor_set.params_dict_from_idx(1)
```

```python
plt.figure(figsize=(20, 5))

for i, img in enumerate(gabor_set.images()):
    plt.subplot(2, 4, i + 1)
    plt.imshow(img, cmap='gray', vmin=-1, vmax=1)
    plt.axis('off')
```
